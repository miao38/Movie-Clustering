Part 1

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\Roy Miao\\Documents\\CS 242\\Projects\\Project #2\\ratings.csv")
new_df = pd.pivot(df, values = "rating", index = "userId", columns = "movieId")
new_df = new_df.fillna(0)
new_df
```

The most common entry in the matrix is 0 and it occurs 5,830,804 times
We call matrices with such properties is called a sparse matrix because it contains mostly 0s

Part 2

```python
#a.
from sklearn import preprocessing

new_df = new_df.T
#mean center the data
X_scaled = preprocessing.scale(new_df, with_std = False)
X_scaled
```

```python
#b.
from sklearn import decomposition
import matplotlib.pyplot as plt


#compute PCA
pca = decomposition.PCA(n_components = 2)
pca.fit(X_scaled)
X_trans = pca.transform(X_scaled)
plt.scatter(X_trans[:, 0], X_trans[:, 1])
plt.title("2.b Plot")
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
plt.show()
```

```python
#c.
var = pca.explained_variance_ratio_
print(var)
plt.scatter([1,2], var)
plt.xlabel("Components")
plt.ylabel("Percent of Variance Exmplained")
plt.title("Variance")
plt.show()
```

About 17.62% of the variance is explained by component 1 and about 4.190% is explained by component 2. You can kinda see the difference in variance between the two components in the plot since the points seem to be further away from the highest density area on the x-axis, component 1, than on the y-axis, component 2

```python
#d.
var = []
for i in range(1, 1000):
    pca = decomposition.PCA(n_components=i)
    pca = pca.fit(X_scaled)
    X_trans = pca.transform(X_scaled)
    var = np.cumsum(pca.explained_variance_ratio_)
    if var[i-1] >= .8:
        print(var)
        print("The number of required components is: " + str(i))
        break
```

154 principle components are needed to explain 80% of the variance of the data. k = 154 is more accurate than k = 2 because k = 2 explaines less and less of the variance once k is over 2. k = 154 can explain more of the variance and we can explain more relationships because of that. However, it is hard to visualize 154 components

Part 3

```python
#a.
from sklearn.cluster import KMeans

k = [2, 4, 8, 16, 32]
inertia = []
for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans = kmeans.fit(new_df)
    inertia.append(kmeans.inertia_)
print(inertia)
plt.plot(k, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Graph")
plt.show()
```

#b.
An appropriate value for k would be 8 because that is where I think the "elbow" of the plot is

#c.
A common characteristic I would expect the movies clustered together to have is that they have the same rating. This type of interpretation is not always possible because in this case, I think it's pretty clear that movies with the same rating will be clustered together. However, for other datasets. It may not be clear what common characteristic the clusters have

Part 4

```python
#a.
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=32)
svd = svd.fit(new_df)
svd = svd.singular_values_
plt.plot(svd)
plt.title("Singular Value Plot")
plt.ylabel("Singular Value")
plt.xlabel("Number of components")
plt.show()
```

```python
#b.
k = [2,4,8,16,32]
explained = []
for i in k:
    svd = TruncatedSVD(n_components=i)
    svd = svd.fit(new_df)  
    explained.append(svd.explained_variance_ratio_.sum())
print(explained)
plt.plot(k, explained)
plt.xlabel("Number of Components")
plt.ylabel("Sum of Explained Variance Ratio")
plt.title("4.b Plot")
plt.show()
```

The inertia values are significantly larger because the explained variance is a ratio so it makes sense why they are all decimals. Also as the k gets bigger, the inertia gets smaller. But this is the opposite for the explained variance ratio. As the k gets bigger, the explained variance ratio also increases. So it seems there is a positive correlation between k and explained vairnace ratio, and a negative correlation between k and inertia.

```python
#c.
svd = TruncatedSVD(n_components=2)
svd = svd.fit(new_df)
transformed = svd.transform(new_df)
print(transformed)
```

```python
#d.
kmeans = KMeans(n_clusters=8)
kmeans = kmeans.fit(new_df)
y_kmeans = kmeans.predict(new_df)
plt.scatter(transformed[:, 0], transformed[:, 1], c=y_kmeans)
plt.title("4.d. Plot")
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
plt.show()
```

This plot and the plot from 2.b look the same. Probably because they are both 2 dimensions, but the one in 2.b is PCA and this one is SVD. As the x-axis increases, it seems like the spread of the points gets wider along the y-axis. Also, most of the points seem to have an x value of under 20.
